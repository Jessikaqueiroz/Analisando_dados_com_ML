Implemente um algoritmo de PCA a partir do zero. Utilize um conjunto de dados de sua escolha e reduza a dimensionalidade para duas dimensões. Visualize os dados originais e os dados transformados.

```python
import numpy as np
import matplotlib.pyplot as plt


def normalizar_dados(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

def pca(X, n_componentes=2):

    X_normalizado = normalizar_dados(X)

    cov_matriz = np.cov(X_normalizado.T)

    autovalores, autovetores = np.linalg.eig(cov_matriz)

    idx = np.argsort(autovalores)[::-1]
    autovalores = autovalores[idx]
    autovetores = autovetores[:, idx]

    componentes_principais = autovetores[:, :n_componentes]

    X_transformado = np.dot(X_normalizado, componentes_principais)

    return X_transformado

np.random.seed(0)
X_exemplo = np.random.rand(100, 5)

X_pca = pca(X_exemplo)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_exemplo[:, 0], X_exemplo[:, 1], color='blue', label='Dados Originais')
plt.title('Dados Originais (2 primeiras características)')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], color='red', label='Dados após PCA')
plt.title('Dados após PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()

plt.tight_layout()
plt.show()
```

Implemente um sistema de recomendação baseado em conteúdo usando um conjunto de dados de produtos (como artigos de e-commerce).

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = {'produto': ['Livro A', 'Livro B', 'Livro C'],
        'descricao': ['Livro de aventura',
                      'História de suspense',
                      'Livro de fantasia']}

df = pd.DataFrame(data)

vetorizador = TfidfVectorizer()
tfidf_matrix = vetorizador.fit_transform(df['descricao'])

similaridade_produtos = cosine_similarity(tfidf_matrix)

print('Matriz de Similaridade entre os Produtos:')
print(similaridade_produtos)

index_produto = 0
similares = list(enumerate(similaridade_produtos[index_produto]))
similares = sorted(similares, key=lambda x: x[1], reverse=True)

print(f'Recomendações para {df["produto"].iloc[index_produto]}:')
for i in similares[1:]:
    print(df['produto'].iloc[i[0]], 'com similaridade:', i[1])
```

Realize a otimização de hiperparâmetros utilizando uma ferramenta de AutoML em um modelo de árvore de decisão ou floresta aleatória. Avalie o desempenho do modelo otimizado.

Implemente o método Bagging (Bootstrap Aggregating) a partir do zero usando árvores de decisão. Compare a precisão do modelo Bagging com uma única árvore de decisão.

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.utils import resample
from sklearn.model_selection import train_test_split

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

def bagging(X_train, y_train, n_estimators=10):
    modelos = []

    for _ in range(n_estimators):
        X_resample, y_resample = resample(X_train, y_train)
        modelo = DecisionTreeClassifier(random_state=42)
        modelo.fit(X_resample, y_resample)
        modelos.append(modelo)

    return modelos

def previsoes_bagging(modelos, X_test):
    previsoes = np.array([modelo.predict(X_test) for modelo in modelos])
    previsoes_agregadas = np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=0, arr=previsoes)
    return previsoes_agregadas

n_estimators = 10
modelos_bagging = bagging(X_train, y_train, n_estimators=n_estimators)

y_pred_bagging = previsoes_bagging(modelos_bagging, X_test)

acuracia_bagging = accuracy_score(y_test, y_pred_bagging)
print(f'Acurácia do modelo Bagging ({n_estimators} estimadores): {acuracia_bagging:.4f}')

modelo_arvore_unica = DecisionTreeClassifier(random_state=42)
modelo_arvore_unica.fit(X_train, y_train)

y_pred_arvore_unica = modelo_arvore_unica.predict(X_test)

acuracia_arvore_unica = accuracy_score(y_test, y_pred_arvore_unica)
print(f'Acurácia da árvore de decisão única: {acuracia_arvore_unica:.4f}')
```

Após aplicar PCA em um conjunto de dados, analise a contribuição de cada componente principal e como isso afeta a interpretação dos dados.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

X_normalizado = StandardScaler().fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_normalizado)

pca_df = pd.DataFrame(data=X_pca, columns=[f'Componente {i+1}' for i in range(X_pca.shape[1])])
pca_df['Classe'] = [target_names[i] for i in y]

variancia_explicada = pca.explained_variance_ratio_

plt.figure(figsize=(8, 6))
plt.bar(range(1, len(variancia_explicada) + 1), variancia_explicada, alpha=0.7, align='center', label='Variância Explicada')
plt.ylabel('Proporção da Variância Explicada')
plt.xlabel('Componentes Principais')
plt.title('Variância Explicada por Cada Componente Principal')
plt.xticks(range(1, len(variancia_explicada) + 1))
plt.axhline(y=0.95, color='r', linestyle='--')
plt.text(1, 0.95, '95% Limite', color='red', ha='center')
plt.legend()
plt.show()

plt.figure(figsize=(8, 6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_name)

plt.title('PCA - Iris')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.grid()
plt.show()
```

Após aplicar LDA em um conjunto de dados, analise como as classes estão separadas e discuta o impacto das variáveis na discriminação das classes.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X_train, y_train)

lda_df = pd.DataFrame(data=X_lda, columns=['Componente 1', 'Componente 2'])
lda_df['Classe'] = [target_names[i] for i in y_train]

plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for i, target_name in enumerate(target_names):
    plt.scatter(lda_df.loc[lda_df['Classe'] == target_name, 'Componente 1'],
                lda_df.loc[lda_df['Classe'] == target_name, 'Componente 2'],
                color=colors[i], alpha=.8, lw=lw, label=target_name)

plt.title('LDA - Separação das Classes no Conjunto de Dados Iris')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.grid()
plt.show()

variancia_explicada = lda.explained_variance_ratio_
print(f'Variância explicada pelos componentes LDA: {variancia_explicada}')
```
